TRY, EXCEPT E FINALLY

```python
x = [1,2,3]
y = 1

try: #tente executar isso, se não executar ele vai printar o erro ali embaixo
    x + y
    x = 1
    y = 1
except:
    print("erro")
```

```python
x = [1,2,3]
y = 1

try: #tente executar isso, se não executar ele vai printar o erro ali embaixo
    x + y
    x = 1
    y = 1
except Exception as e: #trabalhando com o erro
    print("erro {}".format(e))
```

```python
x = [1,2,3]
y = 1

try: #tente executar isso, se não executar ele vai printar o erro ali embaixo
    x + y
    x = 1
    y = 1
except Exception as e: #trabalhando com o erro. Exception é uma cláusula de erro generico, para qualquer erro que aparecer
    print("erro {}".format(e))
finally:
    print("olá")
```
